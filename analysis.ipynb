```python
survey = pd.read_csv('survey.csv')
survey
```

```python
def changeToStateCode(x):
    if (x == 'Virginia') or (x== 'Virginia '):
        return 'VA'
    elif x == 'Maryland ':
        return 'MD'
    elif x == 'New York':
        return 'NY'
    elif x == 'North Carolina ':
        return 'NC'
    elif x == 'California':
        return 'CA'
    elif x == 'Centreville':
        return 'VA'
    else:
        return x
```

```python
def cleanReasons(x):
    if x == 'Passive income and property management ':
        return 'investing'
    elif x == 'I wanna get out this bitch ':
        return 'move to new place'
    elif x == 'Purchased a property June 2020':
        return 'already purchased property'
    else:
        return x
```

```python
surveyLocation = survey['Which state are you located in? '].map(changeToStateCode)
purchaseNow = survey['Whatâ€™s the main reason you want to buy property now?'].map(cleanReasons)
whyNotPurchaseNow = survey['What is the main reason you do not want to buy property in the next 18 months? '].map(cleanReasons)
```

```python
cleanSurvey = pd.DataFrame({'gender':survey['Gender'],'age_group':survey.Age,'location':surveyLocation,
                         'precovid_purchase':survey['Were you considering buying property prior to March 2020?'],
                         'reason_for_not_purchasing_precovid':survey['What is the main reason you did not want to buy property prior to March 2020?'],
                         'purchasing_18months':survey['Are you considering buying property in the next 18 months? '],
                         'reason_to_purchase_now':purchaseNow,'reason_for_not_purchsaing_18months':whyNotPurchaseNow})
cleanSurvey
```

```python
yesPrecovidPurchase = cleanSurvey.loc[cleanSurvey.precovid_purchase == 'Yes']
noPrecovidPurchase = cleanSurvey.loc[cleanSurvey.precovid_purchase == 'No']
```

```python
cleanYesPrecovidPurchase = yesPrecovidPurchase.drop(['reason_for_not_purchasing_precovid','precovid_purchase'],1)
cleanYesPrecovidPurchase
```

```python
cleanNoPrecovidPurchase = noPrecovidPurchase.drop(['precovid_purchase'],1)
cleanNoPrecovidPurchase
```

```python
def counter(x):
    count = {}
    for item in x:
        dictKeys = count.keys()
        if item in dictKeys:
            value = count.get(item)
            value += 1
            count.update({item:value})
        else:
            if item == item:
                count.update({item:1})

    return {k: v for k, v in sorted(count.items(), key=lambda item: item[1],reverse= True)}

def countDF(DF):
    columns = list(DF.columns)
    count = []
    
    for i in columns:
        column = DF[i]
        count.append(counter(column))
        
    return count
```

```python
countedYesPrecovid = countDF(cleanYesPrecovidPurchase)
countedNoPrecovid = countDF(cleanNoPrecovidPurchase)
```

```python
with plt.style.context('seaborn-muted'):
    fig, ax = plt.subplots(nrows=3, ncols=2, figsize=(20,10),constrained_layout=True)
    fig.suptitle('Purchasing Property Prior to March 2020', fontsize=20)
    i = 0
    titles = ['Gender','Age Group','Location','Purchasing Property Within 18 Months?',
              'Why Purchase Property?','Why Not Purchase Property?']
    for row in ax:
        for col in row:
            data = countedYesPrecovid[i]
            col.bar(list(data.keys()),list(data.values()))
            col.title.set_text(titles[i])
            i+=1
```

```python
with plt.style.context('seaborn'):
    fig, ax = plt.subplots(nrows=3, ncols=2, figsize=(20,10),constrained_layout=True)
    fig.suptitle('Not Purchasing Property Prior to March 2020', fontsize=20)
    i = 0
    titles = ['Gender','Age Group','Location','Why Not Purchase Property Prior to March 2020?',
              'Purchasing Property Within 18 Months?','Why Purchase Property?']
    for row in ax:
        for col in row:
            data = countedNoPrecovid[i]
            col.bar(list(data.keys()),list(data.values()))
            col.title.set_text(titles[i])
            col.set_ylabel('Responses')
            i+=1
            
    plt.show()
    
    plt.figure(figsize=(8.25,3))
    plt.bar(list(countedNoPrecovid[-1].keys()),list(countedNoPrecovid[-1].values()))
    plt.title('Why Not Purchase Property?')
    plt.ylabel('Responses')
    plt.show()
```
